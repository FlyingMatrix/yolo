```python
# import dependencies
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
# object detection using YOLOv8n
model = YOLO('yolov8n.pt') 
model.train(data='coco128.yaml', epochs=2) 
```

```python
# predict on an image
result = model.predict(source='https://ultralytics.com/images/bus.jpg', save=False, imgsz=320, conf=0.5)
```

```python
# visualize predicted result
%matplotlib inline
# get the first result and plot it, returns a NumPy array with boxes drawn (BGR format)
pred_img = result[0].plot()  
# convert BGR to RGB, since OpenCV and YOLO use BGR format internally, while Matplotlib expects RGB
pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)
# display inline using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(pred_img)
plt.axis('off')  
plt.show() 
```
